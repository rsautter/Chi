```python
import chi
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

# 1. Synthetic Colored Noise

```python
wn = chi.cNoise(0,[2**13])
rn = chi.cNoise(2,[2**13])
wn = chi.normalize(wn)
rn = chi.normalize(rn)

plt.figure(figsize=(10,5))
plt.plot(wn,label='White Noise',color='blue')
plt.plot(rn,label='Red Noise',color='r')
plt.xlim(0,len(wn))
plt.legend()
plt.show()
```

# 2. Visualizing Set of Singularity Spectrum

```python
wn = chi.cNoise(0,[2**13])
rn = chi.cNoise(2,[2**13])
wn = chi.normalize(wn)
rn = chi.normalize(rn)
ars, fars, cr  = chi.autoMFDFA(rn,nsamples=30)
aws, faws, cw  = chi.autoMFDFA(wn,nsamples=30)


plt.figure(figsize=(10,7))
plt.scatter(aws,faws,color='b',label=r"White Noise $\sigma (\Delta \alpha)$ ="+str(cw))
  

plt.scatter(ars,fars,color='r',label=r"Red Noise $ \sigma (\Delta \alpha)$ ="+str(cr))
  


plt.xlabel(r"$\alpha$",fontsize=20)
plt.ylabel(r"$f(\alpha)$",fontsize=20)

plt.legend()
plt.show()
```

# 3. Visualizing Power Spectrum Density

```python
import numpy as np

psdw = np.fft.fft(wn)
psdw = np.real(psdw*np.conj(psdw))
psdr = np.fft.fft(rn)
psdr = np.real(psdr*np.conj(psdr))

freq = np.fft.fftfreq(len(wn))
seq = np.argsort(freq)


psdw = psdw[seq]
psdr = psdr[seq]
freq = freq[seq]
psdw = psdw[freq>0]
psdr = psdr[freq>0]
freq = freq[freq>0]

plt.figure()
plt.title("White Noise PSD")
plt.loglog(freq,psdw,color="b")
plt.xlabel("Frequency")
plt.ylabel("PSD")

plt.figure()
plt.loglog(freq,psdr,color="r")
plt.title("Red Noise PSD")
plt.xlabel("Frequency")
plt.ylabel("PSD")
plt.show()
```

# 4. $\zeta$ - Space

```python
print("White Noise chi values:")
print(chi.chiSpace(wn))
```

```python
print("Red Noise chi values:")
print(chi.chiSpace(rn))
```

## 4.1. Multiple samples of colored noise

```python
import pandas as pd
import tqdm

sampleSize = 2**13
nsamples = 200
wn,pn,rn = [],[],[]
swn,spn,srn = [],[],[]

for i in tqdm.tqdm(range(nsamples)):
  try:
    serie = chi.cNoise(0,[sampleSize])
    serie = chi.normalize(serie)
    wn.append(chi.chiSpace(serie,nsamples=30))
    swn.append( pickle.dumps(serie))
  except Exception as e:
    # Threshold should be lower to see any Singularity spectrum, but it should not be a problem in this example
    pass
  try:
    serie = chi.cNoise(1,[sampleSize])
    serie = chi.normalize(serie)
    pn.append(chi.chiSpace(serie,nsamples=30) )
    spn.append( pickle.dumps(serie))
  except Exception as e:
    # Threshold should be lower to see any Singularity spectrum, but it should not be a problem in this example
    pass
  try:
    serie = chi.cNoise(2,[sampleSize])
    serie = chi.normalize(serie)
    rn.append(chi.chiSpace(serie,nsamples=30) )
    srn.append( pickle.dumps(serie))
  except Exception as e:
    # Threshold should be lower to see any Singularity spectrum, but it should not be a problem in this example
    pass

wn = pd.DataFrame(wn)
pn = pd.DataFrame(pn)
rn = pd.DataFrame(rn)
wn['serie'] = swn
pn['serie'] = spn
rn['serie'] = srn
```

```python
plt.figure(figsize=(7,7))
plt.scatter(wn["SC"],wn["GQQ"],color='b',label="White Noise")
plt.scatter(pn["SC"],pn["GQQ"],color='magenta',label="Pink Noise")
plt.scatter(rn["SC"],rn["GQQ"],color='r',label="Red Noise")
plt.xlim(0)
#plt.ylim(0,1)
plt.xlabel(r"$S_C$",fontsize=18)
plt.ylabel(r"$G_{QQ}$",fontsize=18)
plt.title(r"$\chi$-Space",fontsize=20)
plt.legend(fontsize=15)
plt.show()
```

```python
wn.to_csv("zWhite.csv")
pn.to_csv("zPink.csv")
rn.to_csv("zRed.csv")
```
